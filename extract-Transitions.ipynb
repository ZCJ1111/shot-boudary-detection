```python
from glob import glob
import matplotlib.pyplot as plt
import numpy as np
import cv2
from videoio import videosave, videoread
from tqdm import tqdm
```

```python
KK = glob('result_MR_lego/*tt*')
KK1 = [int(k.split('_')[-1].split('.')[0]) for k in KK]
np_trans = np.sort(np.array(KK1))
```

```python
idx_start = np.concatenate((np.expand_dims(np.diff(np_trans),0),np.zeros((1,1))),1)>5
idx_end = np.concatenate((np.zeros((1,1)),np.expand_dims(np.diff(np_trans),0)),1)>5
```

```python
kk11 = 0
for start, end in zip(np_trans[np.squeeze(idx_start)], np_trans[np.squeeze(idx_end)]):
    video_np = np.zeros(((end+5)-(start+1),1080, 1920, 3))
    kk10 = 0
    for frm in tqdm(range(start+1, end+5)):
        img_cv = cv2.imread('result_MR_lego/frame_nhe_'+str(frm)+'.png')
        video_np[kk10,:,:,:] = img_cv
        kk10+=1

    videosave("results_shot_lego_bdry/shot_"+str(kk11)+".mp4", video_np/255)
                       
    kk11 += 1
    print(str(kk11)+'/'+str(len(np_trans[np.squeeze(idx_start)])))
```
