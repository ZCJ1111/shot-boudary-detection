```python
from glob import glob
import numpy as np
import os
from videoio import videosave
import matplotlib.pyplot as plt
from tqdm import tqdm
from joblib import Parallel, delayed

if not os.path.exists('/ml_data_fuse/datasets/transition_boundaries_vids/'):
    os.mkdir('/ml_data_fuse/datasets/transition_boundaries_vids/')
    
transition_shot_paths = glob("/ml_data_fuse/datasets/transition_boundaries/*.npz")
transition_shot_paths = ['/ml_data_fuse/datasets/transition_boundaries/220.npz']


def save_video(k):

    for path in tqdm(transition_shot_paths[k:k+1]):
        shot = np.load(path)
        frames = shot["x"]
        
        frame_gap = np.zeros(frames.shape)[:3,:,:,:]
        frame_gap[0,:,:,:] = 255
        frame_gap[1,:,:,:] = 255
        
        frames = np.concatenate((frames,frame_gap),0)
        
    videosave(path.replace('transition_boundaries','transition_boundaries_vids').replace('.npz','.mp4'),frames/255)

save_video(0)
#Parallel(n_jobs=-1)(delayed(save_video)(i) for i in range(len(transition_shot_paths)))
```
